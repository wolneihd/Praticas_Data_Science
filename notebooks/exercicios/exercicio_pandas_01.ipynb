```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregar o CSV: Importe o arquivo sales_data.csv e exiba as primeiras 5 linhas do DataFrame.
df = pd.read_csv('exercicio_pandas_01.csv')
df.head()
```

```python
# Resumo Estatístico: Obtenha um resumo estatístico das colunas Quantity, Price e Total.
df[['Quantity','Price','Total']].describe()
```

```python
# Filtragem Simples: Filtre todas as vendas feitas na categoria Electronics.
df[df['Category'] == 'Electronics']
```

```python
# Vendas por Categoria: Calcule o total de vendas (Total) para cada categoria (Category).
df.pivot_table(values='Total', index='Category', aggfunc='sum')
```

```python
# Produtos Mais Vendidos: Identifique o produto mais vendido (com base na Quantity).
df.pivot_table(values='Quantity', index='Product', aggfunc='sum').sort_values(by='Quantity', ascending=False)
```

```python
# array de valores:
vendas_por_produto = df.pivot_table(values='Quantity', index='Product', aggfunc='sum').sort_values(by='Quantity', ascending=False)
labels = vendas_por_produto.index
vendas = vendas_por_produto['Quantity'].values
```

```python
# plt.style.available
plt.style.use('seaborn-v0_8-pastel')

plt.bar(
    x=labels, 
    height=vendas,
    )

# Adicionar título e rótulos aos eixos
plt.title('Valores')
plt.xlabel('Produtos')
plt.ylabel('Quantidade de vendas')
plt.xticks(rotation=45)

# Exibir o gráfico
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filtrar por Data: Filtre as vendas feitas entre 2024-03-01 e 2024-04-01.
df = pd.read_csv('exercicio_pandas_01.csv', parse_dates=['OrderDate'])
query_01 = df['OrderDate'] > '2024-03-01'
query_02 = df['OrderDate'] < '2024-04-01'
df[query_01 & query_02]
```

```python
# Média por Cidade: Calcule a média de vendas (Total) por cidade.
df.pivot_table(values='Total', index='City', aggfunc='mean').sort_values(by='Total')
```

```python
# Produto Mais Lucrativo: Identifique o produto com o maior valor total de vendas (Total).
df.nlargest(n=1, columns='Total')
```

```python
# Agrupamento e Soma: Agrupe os dados por Cidade (City) e calcule a soma das colunas Quantity e Total.
df.pivot_table(values=['Total','Quantity'], index='City', aggfunc='sum')
```

```python
# Análise Temporal: Crie uma nova coluna Month com base em OrderDate e analise o total de vendas (Total) para cada mês.
df['Month'] = df['OrderDate'].dt.month_name()
```

```python
vendas_por_mes = df.pivot_table(values=['Total'], index='Month', aggfunc='sum')
```

```python
meses = vendas_por_mes.index
vendas = vendas_por_mes['Total'].values
```

```python
plt.style.use('ggplot')

plt.bar(
    x=meses, 
    height=vendas,
    )

# Adicionar título e rótulos aos eixos
plt.title('Valores')
plt.xlabel('Meses')
plt.ylabel('Quantidade de vendas (US$)')
plt.xticks(rotation=45)

# Exibir o gráfico
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
